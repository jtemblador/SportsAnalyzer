## Step 4: Verify data files are accessible

```python
import os

data_path = '/content/drive/MyDrive/SportsAnalyzer/data/nfl'

if not os.path.exists(data_path):
    print(f'ERROR: {data_path} not found')
    print('Check that you uploaded data/nfl/ to My Drive/SportsAnalyzer/data/nfl/')
else:
    print(f'Found data directory: {data_path}\n')
    print('Contents:')
    for folder in sorted(os.listdir(data_path)):
        folder_path = os.path.join(data_path, folder)
        if os.path.isdir(folder_path):
            files = [f for f in os.listdir(folder_path) if f.endswith('.parquet')]
            print(f'  {folder:25s} {len(files)} parquet files')
```

## Step 5: Read a Parquet file to verify data integrity

Loads Patrick Mahomes' 2024 stats as a sanity check.

```python
import pandas as pd

df = pd.read_parquet(f'{data_path}/player_stats/player_stats_2024.parquet')
print(f'Loaded {len(df):,} rows, {len(df.columns)} columns')
print()

# Sanity check: Mahomes 2024 Week 1 passing yards (should be 291)
mahomes = df[(df['player_name'] == 'P.Mahomes') & (df['week'] == 1)]
if not mahomes.empty:
    print(f"Mahomes 2024 W1: {mahomes.iloc[0]['passing_yards']:.0f} passing yards (expected ~291)")
else:
    print('Could not find Mahomes W1 — check data upload')
```
